```python
x_train_tensor = torch.rand(100,6)
y_train_tensor = torch.rand(100,1)
```

# Parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# define loss function
loss_function  = nn.BCELoss()
```

# Training Pipeline

```python
# Create model
model = MySimpleNN(x_train_tensor.shape[1])

# Define Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# Define Loop
for epoch in range(epochs):

  # forward pass
  y_pred = model(x_train_tensor)

  #loss calculate
  loss = loss_function(y_pred, y_train_tensor)

  # clear gradients
  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # parameter update
  optimizer.step()
```
